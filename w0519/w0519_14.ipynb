```python
import pandas as pd
df = pd.read_csv('data/stuscore.csv', index_col='번호')
```

Str 함수

```python
### startswith 함수 - 해당글자로 시작되는 문자 검색
filt = df['이름'].str.startswith('S')
df[filt]
```

```python
### endswith 함수 - 해당글자로 끝나는 문자검색
df[df['이름'].str.endswith('순')]
df[df['이름'].str.endswith('n')]
```

```python
### contains - 해당글자가 포함되어 있는 문자 검색
df[df['이름'].str.contains('s')]
df[df['이름'].str.contains('순')]
```

```python
### a가 들어가 있는 문자를 검색하시오.
df[df['이름'].str.contains('a')].count() # 40개
df[df['이름'].str.contains('A')] # 8개
df[df['이름'].str.contains('A') | df['이름'].str.contains('a')].count() # 48개
```

소문자로 변경후 글자가 있는지 검색

```python
df['이름'].str.lower() # 모든 글자 소문자로 변경

# 모든 글자를 소문자로 변경 후 a가 포함되어 있는 문자 검색
df[df['이름'].str.lower().str.contains('a')].count()

# 모든 글자 대문자로 변경
df['이름'].str.upper() 
df[df['이름'].str.upper().str.contains('A')].count()
```

```python
df = pd.read_excel('data/score.xlsx', index_col="지원번호")

# 홍이 포함되어 있는 학생의 국어 영어 수학 과학 성적을 출력하시오.

filt = df['이름'].str.contains("홍") 
df.loc[filt, ['국어','영어','수학','과학']]

df

# SW 특기가 python, java글자가 포함되어 있는 학생의 이름, 학교, 키, Sw특기를 출력하시오.
filt = df['SW특기'].str.contains("Python")
filt2 = df['SW특기'].str.contains("Java")
df.loc[filt|filt2,['이름','학교','키','SW특기']]

### 3, 리스트 형태로 검색
langs = ['python','java']

## 글자가 일치하는 문자 검색
df[df['SW특기'].str.lower().isin(langs)]
```
