```python
data = {
   '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
   '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
   '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
   '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}

df2 = pd.DataFrame(data)
df2

## 평점이 9.0을 넘는 모든 컬럼을 출력하시오.
## 9.0 미만 영화 이름, 평점 2개 컬럼만 출력하시오

filt = df2['평점']>=9.0
df2[filt]
```

```python
# 키 185이상이면서 수학점수 80점 이상 추력하시오.
filt = df['키']>=185
filt2 = df['수학']>=80
```

```python
df[filt&filt2]
df[~filt&~filt2]
```

```python
df[(df['키']>=185)&(df['수학']>=80)]
df[~(df['키']>=185)&~(df['수학']>=80)]
```

```python
# 키가 185이하 또는 수학이 50점이상
filt = df['키']<=185
filt2 = df['수학']>=50
```

```python
df[filt|filt2]
```

```python
df[(df['키']<=185)|(df['수학']>=50)]
```
