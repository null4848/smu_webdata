```python
import pandas as pd
df = pd.read_excel('data/score.xlsx', index_col='지원번호')
```

```python
### column 출력
df['키']
df[['이름','키']]
df[df.columns[2:5]]
df.columns
```

```python
### row 출력
df[1:5]
df.loc['1번']
df.loc[['1번','2번','3번']]
df.loc['1번':'3번']
df.loc['1번':'3번']['키']
df.loc['1번':'3번'][['키', 'SW특기']]
df.loc['1번':'3번','키':'수학'] ## 슬라이싱 출력
df.iloc[0]
df.iloc[0:3]
df.iloc[[0,3,5]]
df.iloc[0:3][['키','수학']]
df.iloc[0:3,['키','수학']]
```

```python
## 조건식
filt = df['키']>=190
df[filt]
```

```python
df[df['키']>=190]
```

```python
# &, |
```

```python
df[df['키']>=190]
df[(df['키']>=190) & (df['국어']>=90)]
df[(df['키']>=190) | (df['국어']>=90)]
df[~(df['키']>=190)] ### not
```

```python
### str 함수
filt = df['이름'].str.contains('순') # 순이 포함되어 있는 문자
df[filt]

filt = df['이름'].str.endswith('순') # 마지막 글자
df[filt]

filt = df['이름'].str.startswith('홍') # 첫번째 글자
df[filt]
```

```python
arr = ['python','java']
filt = df['SW특기'].isin(arr) # 리스트 타입을 가지고 일치하는 것을 검색
df[filt]
df['SW특기'].str.lower() # 소문자 변경
df['SW특기'].str.upper() # 대문자 변경
filt = df['SW특기'].str.lower().isin(arr)
df[filt]

### java가 포함되어있는 글자도 출력하시오.
# filt = df['SW특기'].str.lower().str.contains('java')
# df[filt]
# df['SW특기'].str.lower().str.contains('java')

# contains -> NaN 데이터 처리를 별도해야 해야함
# na:NaN 데이터 처리 False 검색되지 않음, True 검색됨
df['SW특기'].str.contains('java', na=True)
filt = df['SW특기'].str.lower().str.contains('java',na=False)
df[filt]
filt = df['SW특기'].str.lower().str.contains('java',na=True)
df[filt]

df.info()
```
