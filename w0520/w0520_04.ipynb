데이터 수정

```python
### 디지털고를 소프트고로 변경
# ### 특정 데이터 찾아서 변경
# df['학교'].replace({"디지털고":"소프트고"}, inplace=True)

## 2개를 변경
df['학교'].replace({"가산고":"금천고","디지털고":"단지고"})

## 특정 글자 소문자로 변경
# df['SW특기'] = df['SW특기'].str.lower()

## 특정 글자 대문자로 변경
df['SW특기'] = df['SW특기'].str.upper()

## 첫글자 대문자 나머지 소문자
# df['SW특기'] = df['SW특기'].str.capitalize()

# 학교 -> 가산고등학교, 디지털고등학교
df['학교'] = df['학교'] + '등학교'
df
```

```python
## 컬럼추가
## 없는 컬럼을 입력하면 컬럼이 추가됨
df['합계'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']

## 평균을 추가하시오.
df['평균'] = df['합계']/5
df

# 결과 컬럼을 추가하고 내용은 불합격 입력하시오.
df['결과'] = "불합격"
df

## 평균 컬럼이 60점 이상인 학생만 출력하시오.
# filt = df['평균'] >= 60
# df[filt]

## 평균이 60점 이상이면 결과 커럼에 합격이라고 입력하세요.
# filt = df['평균']>=60
df.loc[filt,'결과'] = " 합격"
df
```
