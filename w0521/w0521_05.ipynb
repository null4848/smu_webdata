```python
import pandas as pd
import numpy as np

data = {
    '번호':['1번','2번','3번','4번','5번','6번','7번','8번'],
    '이름':['홍길동','유관순','이순신','강감찬','김구','김유신','홍길자','홍길순'],
    '달러':['$19,700','$18,400','$16,800','$18,700','$18,800','$20,200','$18,800','$19,000'],
    '우대':[90,40,80,40,15,80,55,100]
}
```

```python
df = pd.DataFrame(data)
df
df.set_index('번호')
df

```

```python
df['금액'] = 0
money = df['달러'].str.replace("$","").str.replace(",","")
money
df['금액'] = money.astype(int) + (money.astype(int) * df['우대'])
df

# 우대ㅔ 50인사람만 저굥해서금액을 산출
#  우대가 50% 달려익력ㅏ라햇고 공ㅇ개줌은 먹지않앗다.

filt = df['우대']<50
df.loc[filt,'결과'] = df['달러']
df.loc[~(filt), '결과'] = money.astype(int) + (money.astype(int) * df['우대'])
df
```

```python
def change(data):
    if data['우대']>=50: return data['달러'] + data['달러'] * data['우대']*0.01
    else: return data['달러']
    
df['금액'] = df.apply(change)
df
```
