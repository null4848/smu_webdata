```python
# intㅌ ㅏ입에 cm 문자 추가방법
# df['키'] = df['키'].astype(str)+'cm'

# 함수 사용해서 추가
# def add_cm(height):
#     return str(height)+'cm'

# df['키'] = df['키'].apply(add_cm)
# df

df['합계'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
df['평균'] = df['합계']/5
df['등급'] = 'F'

# filt = df['평균']>=90
# df.loc[filt,'등급'] = "합격"


# 함수를 사용해서
# 90점 이상 합격, 90점 미만은 불합격
def rank(score):
    result = ""
    if score >= 90: result = "합격"
    else: result = "불합격"
    return result

df['등급'] = df['평균'].apply(rank)
df
```

```python
### 90점 이상 A 80점 이상 B 70점 이상 C 60점이상 D 60점미만 F

def rank(score):
    result = ""
    if score >= 90: result = "A"
    elif score >= 80: result = "B"
    elif score >= 70: result = "C"
    elif score >= 60: result = "D"
    else: result = "F"
    return result


df['등급'] = df['평균'].apply(rank)
df
```

```python
df['요일'] = [0,1,2,3,4,5,3,6]
df
```

```python
## 함수를 이용해서 요일 0: 일 월 화 수 목 금 토 변경하시오
def day(num):
    result = ""
    if num == 0:
        result = "일"
    elif num == 1:
        result = "월"
    elif num == 2:
        result = "화"
    elif num == 3:
        result = "수"
    elif num == 4:
        result = "목"
    elif num == 5:
        result = "금"
    elif num == 6:
        result = "토"
        
    return result
        
df['요일'].apply(day)
df
   
```
